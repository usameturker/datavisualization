# Function Method and Object Oriented Methods

### Function Method

```python
age = np.arange(25, 44, 2.5)
salary = np.arange(40000, 76000, 5000)
salary_2 = [45372, 52345, 54148, 94256, 16544, 45600, 96542, 98148] 
```

```python
plt.style.use("seaborn-darkgrid")

plt.title("Salary By Age")
plt.ylabel("Salary")
plt.xlabel("Age")
#plt.grid(color="purple") #gridleri renklendirmek için kullanılabilir
plt.plot(age, salary);
```

### Object Oriented Methods (OOM)

```python
fig, ax = plt.subplots()

ax.plot(age, salary)
ax.set_xlabel("Age")
ax.set_ylabel("Salary")
ax.set_title("Salary By Age");
```

### Functional

```python
plt.subplot(2, 1, 1)
plt.plot(age, salary, "r")

plt.subplot(2, 1, 2)
plt.plot(age, salary_2, "b");
```

```python
plt.subplot(2, 1, 2)
plt.plot(age, salary, "r")

plt.subplot(2, 1, 1)
plt.plot(age, salary_2, "b");
```

### Object Oriented Methods (OOM)

```python
fig, ax = plt.subplots(2, 1)

ax[0].plot(age, salary, "r")
ax[1].plot(age, salary_2, "b")

plt.tight_layout();
```

```python
fig, ax = plt.subplots(2, 4, figsize=(12,6))

ax[0,0].plot(age, salary, "r")
ax[0,1].plot(age, salary_2, "b")
ax[0,2].plot(age, salary, "r")
ax[0,3].plot(age, salary_2, "b")
ax[1,0].plot(age, salary, "r")
ax[1,1].plot(age, salary_2, "b")
ax[1,2].plot(age, salary, "r")
ax[1,3].plot(age, salary_2, "b")
plt.tight_layout();
```

```python
fig, ax = plt.subplots(4, 2, figsize=(12,6))

ax[0,0].plot(age, salary, "r")
ax[0,1].plot(age, salary_2, "b")
ax[1,0].plot(age, salary, "r")
ax[1,1].plot(age, salary_2, "b")
ax[2,0].plot(age, salary, "r")
ax[2,1].plot(age, salary_2, "b")
ax[3,0].plot(age, salary, "r")
ax[3,1].plot(age, salary_2, "b")
plt.tight_layout();
```

### Functional

```python
plt.subplot(2, 3, 1)
plt.plot(age, salary, "r")

plt.subplot(2, 3, 2)
plt.plot(age, salary, "r")

plt.subplot(2, 3, 3)
plt.plot(age, salary, "r")

plt.subplot(2, 3, 4)
plt.plot(age, salary, "b")

plt.subplot(2, 3, 5)
plt.plot(age, salary, "b")

plt.subplot(2, 3, 6)
plt.plot(age, salary, "b")

plt.tight_layout();
```

### OOM

```python
fig, ax = plt.subplots(2,3)

ax[0][0].plot(age, salary, "r")
ax[0][1].plot(age, salary, "r")
ax[0][2].plot(age, salary, "r")

ax[1][0].plot(age, salary, "b")
ax[1][1].plot(age, salary, "b")
ax[1][2].plot(age, salary, "b")

plt.tight_layout();
```

```python
x = np.arange(0, 30, 0.2)

fig, ax = plt.subplots(figsize = (14, 8))

ax.plot(x, x-1, color = "green", linewidth= 3, linestyle = "-")
ax.plot(x, x-2, color = "red", lw= 3, ls = "-.")
ax.plot(x, x-3, color = "yellow", lw= 3, ls = ":")
ax.plot(x, x-4, color = "blue", lw= 3, ls = "--");
```

```python
x = np.arange(0, 10)

fig, ax = plt.subplots(2, 2, figsize = (12, 6))

plt.suptitle("LineStyle & Markers & Markersize & Color", fontsize = 16, y=0.93)#y=y ekseninde hareket ettirir

ax[0, 0].plot(x, "r", marker = "+", markersize = 5)
ax[0, 1].plot(x-2, "y", marker = "o", ms = 5)
ax[1, 0].plot(x-3, "black", marker = "1", ms = 5)
ax[1, 1].plot(x-4, "g", marker = "s", ms = 5)

plt.show()
```

### Multiple plots

```python
x = np.arange(-50, 50)
y1 = np.power(x, 2)
y2 = np.power(x, 3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)



fig1, ax1 = plt.subplots(nrows = 3, ncols = 3, figsize= (20, 20))

ax1[0, 0].plot(x, y1, "tab:blue") # set the color of line chart
ax1[0, 0].set_title("Square Function") # setting title of subplot
ax1[0, 0].set_xlabel("X", fontsize = 18)
ax1[0, 0].set_ylabel("Y", fontsize = 18)

ax1[0, 1].plot(x, y2, "tab:orange") # set the color of line chart
ax1[0, 1].set_title("Cubic Function") # setting title of subplot
ax1[0, 1].set_xlabel("X", fontsize = 18)
ax1[0, 1].set_ylabel("Y", fontsize = 18)

ax1[0, 2].plot(x, y3, "tab:green") # set the color of line chart
ax1[0, 2].set_title("Sine Function") # setting title of subplot
ax1[0, 2].set_xlabel("X", fontsize = 18)
ax1[0, 2].set_ylabel("Y", fontsize = 18)

ax1[1, 0].plot(x, y4, "r-") # set the color of line chart
ax1[1, 0].set_title("Cosine Function") # setting title of subplot
ax1[1, 0].set_xlabel("X", fontsize = 18)
ax1[1, 0].set_ylabel("Y", fontsize = 18)

ax1[1, 1].plot(x, y5, "b-") # set the color of line chart
ax1[1, 1].set_title("Tangent Function") # setting title of subplot
ax1[1, 1].set_xlabel("X", fontsize = 18)
ax1[1, 1].set_ylabel("Y", fontsize = 18)

ax1[1, 2].plot(x, y6, "g-") # set the color of line chart
ax1[1, 2].set_title("Hyperbolic Tangent Function") # setting title of subplot
ax1[1, 2].set_xlabel("X", fontsize = 18)
ax1[1, 2].set_ylabel("Y", fontsize = 18)

ax1[2, 0].plot(x, y7, "m-") # set the color of line chart
ax1[2, 0].set_title("Hyperbolic Sine Function") # setting title of subplot
ax1[2, 0].set_xlabel("X", fontsize = 18)
ax1[2, 0].set_ylabel("Y", fontsize = 18)

ax1[2, 1].plot(x, y8, "y*") # set the color of line chart
ax1[2, 1].set_title("Hyperbolic Cosine Function") # setting title of subplot
ax1[2, 1].set_xlabel("X", fontsize = 18)
ax1[2, 1].set_ylabel("Y", fontsize = 18)

ax1[2, 2].plot(x, y9, "kv") # set the color of line chart
ax1[2, 2].set_title("Exp Function") # setting title of subplot
ax1[2, 2].set_xlabel("X", fontsize = 18)
ax1[2, 2].set_ylabel("Y", fontsize = 18)

plt.tight_layout();
plt.show()
```

## Pie Chart

```python
x = np.array([10, 10, 10, 10])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]


plt.figure(figsize=(10, 8))
plt.pie(x, labels = mylabels, labeldistance=0.50,autopct ="%.1f")

plt.show()
```

```python
x = np.array([100, 100, 100, 100])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode= [0.1, 0, 0, 0]

plt.figure(figsize=(10, 8))
plt.pie(x, labels = mylabels, labeldistance=0.50,autopct ="%.2f", explode=myexplode)

plt.show()
```

```python
plt.figure(figsize=(10,8))

color_list= ["gold", "yellowgreen", "lightcoral", "lightskyblue", "lightgreen", "pink"]

explode_list= [0.1, 0.1, 0.1, 0.1]

y = np.array([25, 35, 25, 15])

mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels,  labeldistance=0.75, autopct = "%.1f", startangle = 90, shadow = True, pctdistance=0.6, 
        colors= color_list, explode= explode_list)

plt.legend(loc = (0));
```
