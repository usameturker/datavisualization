```python
tips=sns.load_dataset('tips')
tips
```

```python
tips_gr=tips.groupby(['size'])
tips_gr['tip'].value_counts()
```

```python
tips['size'].unique()
```

```python
sns.scatterplot(data=tips,x='total_bill',y='tip');
```

```python
sns.regplot(tips,x=tips.total_bill,y=tips.tip)
plt.show()
```

```python
plt.figure(figsize = (10, 5))
sns.scatterplot(x = tips["total_bill"], y = tips["tip"]);
```

```python
plt.figure(figsize = (10, 5))
sns.scatterplot(x ="day", y = "tip", data = tips, hue ="sex", style='smoker');
```

```python
g = sns.FacetGrid(tips, col = "sex", hue = "smoker")
g.map(sns.scatterplot, "total_bill", "tip", alpha = .5)
g.add_legend();
```

```python
g = sns.FacetGrid(tips, col = "sex", hue = "smoker")
g.map(sns.kdeplot, "total_bill", "tip")
g.add_legend();
```

```python
sns.displot(data=tips,x='total_bill',kind='hist')
sns.displot(data=tips,x='total_bill',kind='ecdf')
sns.displot(data = tips, x = "total_bill", kind = "kde");
```

```python
sns.displot(data = tips, x = "total_bill", kind = "hist", kde = True, aspect=5);
```
